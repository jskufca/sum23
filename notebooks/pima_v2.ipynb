```python

#import pandas
import pandas as pd
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
# load dataset
# pima = pd.read_csv("diabetes.csv", header=None, names=col_names, skiprows=1)
pima = pd.read_csv("../data/diabetes.csv", header=None, names=col_names, skiprows=1)

pima.head()
```

```python

#split dataset in features and target variable
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
X = pima[feature_cols] # Features
y = pima.label # Target variable

```

```python

# split X and y into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=16)
```

```python

# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(random_state=16,max_iter=300)

# fit the model with data
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)
```

```python

# import the metrics class
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python

# import required modules
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python

from sklearn.metrics import classification_report
target_names = ['without diabetes', 'with diabetes']
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python

y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python

ypred_t=(y_pred_proba>.32596)+0
cnf_matrix_n = metrics.confusion_matrix(y_test, ypred_t,normalize=('true'))
cnf_matrix= metrics.confusion_matrix(y_test, ypred_t)



class_names=[0,1] # name  of classes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g', ax = ax1)
ax1.set_title('Counts')
ax1.set_xlabel('PREDICTED VALUES')
ax1.set_ylabel('ACTUAL VALUES')

sns.heatmap(pd.DataFrame(cnf_matrix_n), annot=True, cmap="YlGnBu" ,fmt='g', ax = ax2)
ax2.set_title('Row Normalized')
ax2.set_xlabel('PREDICTED VALUES')
ax2.set_ylabel('ACTUAL VALUES')
plt.show()


```
